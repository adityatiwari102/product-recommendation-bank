```python
x = data.groupby('User_id').count()['Ratings'] >1
user_with_ratings = x[x].index
```

```python
filtered_rating = data[data['User_id'].isin(user_with_ratings)]
```

```python
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Product').count()['Ratings']>=10
popular_product = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Product'].isin(popular_product)]
```

```python
final_ratings
```

```python
prod_final = final_ratings.pivot_table(index= ['Status', 'Product'], columns= 'User_id', values='Ratings')
```

```python
prod_final
```

```python
prod_final.fillna(0,inplace=True)
```

```python
prod_final
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_score = cosine_similarity(prod_final)
```

```python
np.where(prod_final.index == ('ACTIVE', 'Home Loan'))[0][0]
```

```python
def recommend(status, product):
    
    index = np.where(prod_final.index == (status, product))[0][0]
    similar_products = sorted(list(enumerate(similarity_score[index])), key= lambda x: x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_products:
        products = []
        products.append(prod_final.index[i[0]])
        
        data.append(products)
        
    return data
```

```python
a = recommend('ACTIVE','Car Loan')
```
